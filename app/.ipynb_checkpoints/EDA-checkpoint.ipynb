# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
data = pd.read_csv('property24_data.csv')
```

```python
data
```

## Dropping Duplicates

```python
data = data.drop_duplicates()
data
```

## Overview of the data

```python
data.dtypes.value_counts().plot.pie()
data.dtypes.value_counts()
```

## Converting data

```python
df = data.copy()
df
```

```python
# Changing the type to float
df['Erf Size'] = data['Erf Size'].str.strip('m²').str.replace(' ','')
df['Floor Size'] = data['Floor Size'].str.strip('m²').str.replace(' ','')
df['Erf Size'] = df['Erf Size'].str.replace('ha','0000')
df
```

## Deleting some columns

```python
df = df.drop(['Listing Number', 'Listing Date', 'Town'], axis=1)
df
```

# Data visualization

```python
df.info()
```

```python
df['Erf Size'] = df['Erf Size'].astype(float)
df['Floor Size'] = df['Floor Size'].astype(float)
```

```python
df.describe()
```

```python
no_binary = df[['Bathrooms', 'Bedrooms', 'Erf Size', 'Floor Size', 'Garages', 'Price', 'Secure Parking', 'Parking']]
```

```python

for col in no_binary.select_dtypes('float'):
    plt.figure(figsize = (12,8))
    plt.title(col)
    sns.boxplot(data = df[col])
    
```

```python
df['Type of Property'].value_counts().plot.pie()
```

## Visualization towards the Price column

```python
for col in no_binary.select_dtypes('float'):
    plt.figure(figsize = (10, 8))
    plt.scatter(no_binary[col], df.Price)
    plt.xlabel(col)
    plt.ylabel('Price')
```

```python
sns.stripplot(x='Type of Property', y="Price",jitter=False, data=df)
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(no_binary.corr(), annot = True)
```

## Analysis
* There are many outliers which are not coherent like max Erf Size is 9.490000e+07
* min Erf Size is a negative value
* There are samples with 20+ bedrooms but have a lower Price
* A house is more expensive than a Townhouse or a Flat
* Rooms and Floor Size are high correlated with price
* The correlation between Erf Size and Price is weird, maybe because af the outliers mentionned earlier

Let's remove those outliers and replot the correlation map

```python
no_binary[no_binary['Erf Size']<0]
```

```python
df_pre = no_binary.copy()
df_pre.describe()
```

```python
df_pre = df_pre[(df_pre['Erf Size']<50000)]
```

```python
df_pre.describe()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(df_pre.corr(), annot = True)
```

## Conclusion

* Now, Erf Size and Price have a better correlation
* Outliers which are not coherent can really affect the data in bad ways so we have to clean them in preparation
